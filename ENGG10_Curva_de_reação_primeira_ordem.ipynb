## Modelo de primeira ordem

\begin{align}
\tau_p \frac{dh(t)}{dt} + h(t) = K_p q_i(t)
\end{align}

```python
import numpy as np
import plotly.graph_objects as go
import pandas as pd
sheet_url="https://docs.google.com/spreadsheets/d/1UPq-_KwH0DZXkwSryfIc5ePeHTOPsQVSuMVaHeimwSI/edit#gid=0"
url_1=sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')

arquivo=pd.read_csv(url_1,decimal=',')
arquivo.head()

tsim = np.array(arquivo['t'],dtype='float64')
Q = np.array(arquivo['Qi(t)'],dtype='float64')
H = np.array(arquivo['H(t)'],dtype='float64')

trace1 = go.Scatter(x=tsim, y=H, name='H(t)', yaxis='y1')
trace2 = go.Scatter(x=tsim, y=Q, name='Q(t)', yaxis='y2')
layout = go.Layout(yaxis=dict(title='H(t)',side='left'),
                   yaxis2=dict(title='Q(t)',overlaying='y',side='right'))

fig = go.Figure(data=[trace1,trace2],layout=layout)
fig.show()
```
